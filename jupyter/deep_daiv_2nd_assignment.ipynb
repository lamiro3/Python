```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('한국표준과학연구원_대표홈페이지_논문_공개목록_10_14_2021.csv', index_col = 0)
```

```python
df.head(10)
```

```python
df[df.duplicated(subset = '논문제목')] # 중복값 확인
```

```python
df.drop_duplicates(subset = '논문제목', keep = 'first', inplace = True) # 중복 요소 제거
```

```python
df.reset_index(drop = True, inplace = True) # 인덱스 재설정
```

```python
vectorizer = TfidfVectorizer(min_df=1, ngram_range=(1,1))
features = vectorizer.fit_transform(df['논문제목'])
print(features)
```

```python
feature_names = vectorizer.get_feature_names()
# print(feature_names)
```

```python
dtm_np = np.array(features.todense())
```

```python
pd.DataFrame(data = dtm_np, columns = feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(dtm_np, dtm_np)
```

```python
indices = pd.Series(df.index, index=df['논문제목'])
```

```python
set(df['논문제목'])
```

```python
idx = indices['Changes in the Photoluminescence of Monolayer and Bilayer Molybdenum Disulfide during Laser Irradiation']
```

```python
sim_scores = list(enumerate(cosine_sim[idx]))
```

```python
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
```

```python
sim_scores = sim_scores[1:51]
```

```python
paper_indices = [i[0] for i in sim_scores]
```
